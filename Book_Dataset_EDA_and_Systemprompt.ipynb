```python
def generate_input_output_pair(prompt, target_responses):
  chat_templates = tokenizer.apply_chat_template(prompt, continue_final_message= True, tokenize= False)
  # print(chat_templates)
  full_response_text = [
      (chat_template + " " + target_response + tokenizer.eos_token)
      for chat_template, target_response in zip(chat_templates, target_responses)
      ]
  # print(full_response_text)

  input_ids_tokenized = tokenizer(full_response_text, add_special_tokens = False, return_tensors ="pt")["input_ids"]
  label_tokenized = tokenizer([" " + response + tokenizer.eos_token for response in target_responses], add_special_tokens = False,
                              return_tensors ="pt", padding = "max_length", max_length = input_ids_tokenized.shape[1])["input_ids"]

  label_tokenized_fixed = torch.where(label_tokenized != tokenizer.pad_token_id, label_tokenized, -100)
  label_tokenized_fixed[:, -1] = tokenizer.eos_token_id

  input_ids_tokenized_left_shifted = input_ids_tokenized[:, :-1]
  label_tokenized_right_shifted = label_tokenized_fixed[:, 1:]

  attention_mask = input_ids_tokenized_left_shifted != tokenizer.pad_token_id
  return {
      "input_ids" : input_ids_tokenized_left_shifted,
      "attention_mask" : attention_mask,
      "labels" : label_tokenized_right_shifted
  }
```

```python
prompt = [
    {"role" : "user", "content" : "Capital of Nepal?"},
    {"role" : "assistant", "content" : "Capital:"}
],
target_responses = ["Kathmandu"]
```

```python
import torch.nn as nn

def calculate_loss(logits, labels):
  loss_fn = nn.CrossEntropyLoss(reduction= 'none')
  entropyloss = loss_fn(logits.view(-1, logits.size(-1)), labels.view(-1))
  return entropyloss
```

# PEFT: LORA Adaptor

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    task_type = "CAUSAL_LM",
    r = 64,
    lora_alpha = 16,
    lora_dropout = 0.1,
    target_modules = ['q_proj', 'v_proj']
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

# Training loop

```python
training_prompt = [
    {
        "role" : "user", "content" : "Who to subscribe for travel videos?"
    },
    {
        "role" : "assistant", "content" : "Subscribe to"
    }
]
target_response = ["TravelWithRam"]
```

```python
data = generate_input_output_pair(prompt= [training_prompt], target_responses= target_response)
```

```python
data["input_ids"] = data["input_ids"].to(device)
```

```python
data["input_ids"].shape
```

```python
out = model(input_ids = data["input_ids"].to(device))
```

```python
out.logits.view(-1, out.logits.size(-1)).shape
```

```python
data["labels"] = data["labels"].to(device)
```

```python
data["labels"].view(-1).shape
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr = 1e-3, weight_decay= 0.01)

for i in range(10):
  out = model(input_ids = data["input_ids"].to(device))
  print(out.logits.shape)
  loss = calculate_loss(out.logits, data["labels"]).mean()

  loss.backward()
  optimizer.step()
  optimizer.zero_grad()

  print(f"loss : {loss.item()}")
```

```python
test_tokenized = tokenizer.apply_chat_template(training_prompt, continue_final_message= True, return_tensors ="pt").to(device)
test_out = model.generate(test_tokenized, max_new_tokens = 10)
print(tokenizer.batch_decode(test_out, skip_special_tokens= True)[0])
```
